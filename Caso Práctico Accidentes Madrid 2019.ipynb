```python
def describe_columna(df, col):
    print(f'Columna: {col} - Tipo de datos: {df[col].dtype}')
    print(f'Número de valores nulos: {df[col].isnull().sum()} - Número de valores distintos: {df[col].nunique()}')
    print('Valores más frecuentes:')
    for i, v in df[col].value_counts().iloc[:10].items() :
        print(i,'\t', v)   
```

```python
accidentes_new.loc[accidentes_new['Estado metereológico'] == 'Se desconoce', 'Estado metereológico'] = np.NaN
# todos los observaciones(filas) de las variables(columnas) Estado meteorológico y Rango de Edad que se que se correspondan con 'Se desconoce' o 'DESCONOCIDA' respectivamente, 
# los convertimos a valores nulos
accidentes_new.loc[accidentes_new['Rango edad'] == 'DESCONOCIDA', 'Rango edad'] = np.NaN
```

```python
accidentes_new.info()
```

```python
rangos_edad = pd.CategoricalDtype(
    categories =['DE 0 A 5 AÑOS', 'DE 6 A 9 AÑOS', 'DE 10 A 14 AÑOS', 'DE 15 A 17 AÑOS',
     'DE 18 A 20 AÑOS', 'DE 21 A 24 AÑOS', 'DE 25 A 29 AÑOS', 'DE 30 A 34 AÑOS', 
     'DE 35 A 39 AÑOS', 'DE 40 A 44 AÑOS', 'DE 45 A 49 AÑOS', 'DE 50 A 54 AÑOS',
     'DE 55 A 59 AÑOS', 'DE 60 A 64 AÑOS', 'DE 65 A 69 AÑOS', 'DE 70 A 74 AÑOS', 
     'MAYOR DE 74 AÑOS'],
    ordered=True)

accidentes_new['Rango edad'] = accidentes_new['Rango edad'].astype(rangos_edad)
```

```python
describe_columna(accidentes, 'LESIVIDAD*')
```

```python
accidentes_new['Lesividad'] = accidentes['LESIVIDAD*'] # creamos la nueva columna Lesividad a partir de la original
```

```python
# una vez más vamos a transformar valores, en este caso los valores en blanco los vamos a codificar con valor 14, porque según el documento indican lo mismo, "Sin asistencia 
# sanitaria"
accidentes_new.loc[accidentes_new['Lesividad'].isnull(), 'Lesividad'] = 14
accidentes_new.loc[accidentes_new['Lesividad'] == 77, 'Lesividad'] = np.NaN
# los campos con valor 77 los transformamos a nulos, porque según el documento indican que es una información desconocida
```

```python
# al igual que con rangos_edad, definimos un nuevo tipo categórico, con su correspondiente orden
grados_gravedad = pd.CategoricalDtype(['Ileso','Leve', 'Grave','Fallecido'], ordered=True) 

#creamos un diccionario de datos que mapee los valores de lesividad y los asocie a los de gravedad
dict_gravedad = {14:'Ileso', 3:'Grave', 4:'Fallecido'}

# Creamos una nueva columna basándonos en la columna Lesividad cuyos valores ya hemos transformado acorde a la interpretación de documento. Sobre esta nueva columna vamos a
# aplicar una función lambda (funciones anónimas, para un sólo uso), para que asocie los valores siguiendo el criterio del diccionario de datos 'dict_gravedad' 
# y por defecto asocie a todos los valores no definidos la categóría 'Leve')
accidentes_new['Gravedad'] = accidentes_new['Lesividad'].apply(
        lambda x: dict_gravedad.get(x, 'Leve')).astype(grados_gravedad) # por último, con 'astype' transformamos la columna a tipo categórico de modo que utilice el orden de 
# las categorías previamente creadas en la variable grados_gravedad
```

```python
top10_vehiculos = accidentes_new['Tipo vehículo'].value_counts()[:10].index.tolist() # index.list indexa unos valores y los devuelve en forma de lista
                                # recordemos que value.counts()recoge los valores más frecuentes, en este caso [:10] nos quedamos con los 10 primeros
type(top10_vehiculos)
```

```python
accidentes_new.groupby(['Tipo vehículo', 'Tipo persona'])['Nº Expediente'].count().unstack().loc[top10_vehiculos]
```

```python
acc_por_sexo = accidentes_new.groupby(['Tipo vehículo', 'Sexo'])['Nº Expediente'].count().unstack().loc[top10_vehiculos]

acc_por_sexo.apply(lambda x: x/x.sum(), axis=1).style.format("{:.1%}")
```

```python
import altair as alt

data = accidentes_new.groupby(['Rango edad', 'Sexo'])['Nº Expediente'].count().reset_index()

alt.Chart(data, width=300, height=150, title='Accidentes en Madrid durante 2019').mark_bar().encode(
    x = alt.X('Rango edad:O', sort = rangos_edad.categories.to_list(), title = None),
    y = alt.Y('Nº Expediente:Q', title = 'Numero de implicados'),
    facet=alt.Facet('Sexo:N', title = None))
```
